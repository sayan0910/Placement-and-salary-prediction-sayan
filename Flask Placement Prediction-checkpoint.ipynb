```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.feature_selection import mutual_info_classif
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"Placement_Data.csv")
```

```python
data
```

```python

X=data.drop(columns=['status','salary','sl_no','specialisation'])

```

```python
labelencoder = LabelEncoder()
Y=data['status']
Y = labelencoder.fit_transform(Y)

```

```python
Y
```

```python
X
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
one=OneHotEncoder()
```

```python
one.fit(X[['gender','ssc_b','hsc_b','hsc_s','degree_t','workex']])
```

```python
X['gender']
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
column_trans=make_column_transformer((OneHotEncoder(categories=one.categories),['gender','ssc_b','hsc_b','hsc_s','degree_t','workex']),remainder='passthrough')
```

```python
column_trans
```

```python
X['gender']
```

```python
classifier = KNeighborsClassifier(weights='distance')
```

```python
X
```

```python
pipe=make_pipeline(column_trans,classifier)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
import pickle
```

```python
pickle.dump(pipe,open('KnnClassificationFlask.pickle','wb'))
```
